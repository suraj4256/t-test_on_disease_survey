```python

df_exploded = df.assign(
    DISEASE=df['Diseases'].str.split(',')
).explode('DISEASE')

df_exploded['DISEASE'] = df_exploded['DISEASE'].str.strip()
df_exploded = df_exploded[df_exploded['DISEASE'].notna() & (df_exploded['DISEASE'] != '')]
df_exploded['DISEASE'] = df_exploded['DISEASE'].replace(disease_mapping)
```

```python
df_exploded.head()
```

```python
df_exploded['LOCALITY'] = df_exploded['Address'].astype(str).str.upper().str.strip()
```

```python
# Step 6: Count affected persons per disease per locality
counts = df_exploded.groupby(['DISEASE', 'LOCALITY'])['Name'].count().reset_index()
counts = counts.rename(columns={'Name': 'Affected Persons'})
```

```python
# Step 7: Total affected per disease
totals = counts.groupby('DISEASE')['Affected Persons'].sum().reset_index()
totals = totals.rename(columns={'Affected Persons': 'Total Affected'})
```

```python
# Step 8: Merge and calculate % affected
merged = pd.merge(counts, totals, on='DISEASE')
merged['% Affected'] = (merged['Affected Persons'] / merged['Total Affected']) * 100
```

```python
# Step 9: Summary stats: Average %, SD, SE
summary = merged.groupby('DISEASE')['% Affected'].agg(
    Average_Percent='mean',
    SE=lambda x: x.std() / np.sqrt(len(x))
).reset_index()
```

```python
summary
```

```python
# Step 10: Merge summary back
final = pd.merge(merged, summary, on='DISEASE')
```

```python
final
```

```python
# Per-locality t-score (optional — comment out if not needed)
# final['T-Bar'] = (final['% Affected'] - final['Average_Percent']) / final['SE']

# Compute T-bar (overall t-statistic for the disease)
summary['T-bar'] = summary['Average_Percent'] / summary['SE']

# Merge T-bar into final table
final = pd.merge(final, summary[['DISEASE', 'T-bar']], on='DISEASE')
```
